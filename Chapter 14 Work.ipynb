```python
import sqlite3

# Function to create the database and table
def create_phonebook_db():
    conn = sqlite3.connect("phonebook.db")
    cursor = conn.cursor()
    cursor.execute('''
        CREATE TABLE IF NOT EXISTS Entries (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            name TEXT NOT NULL,
            phone_number TEXT NOT NULL
        )
    ''')
    conn.commit()
    conn.close()

# Function to add a new entry to the phonebook
def add_entry(name, phone_number):
    conn = sqlite3.connect("phonebook.db")
    cursor = conn.cursor()
    cursor.execute("INSERT INTO Entries (name, phone_number) VALUES (?, ?)", (name, phone_number))
    conn.commit()
    conn.close()

# Function to look up a person's phone number
def lookup_entry(name):
    conn = sqlite3.connect("phonebook.db")
    cursor = conn.cursor()
    cursor.execute("SELECT phone_number FROM Entries WHERE name=?", (name,))
    result = cursor.fetchone()
    conn.close()
    if result:
        return result[0]
    else:
        return "Entry not found."

# Function to change a person's phone number
def change_entry(name, new_phone_number):
    conn = sqlite3.connect("phonebook.db")
    cursor = conn.cursor()
    cursor.execute("UPDATE Entries SET phone_number=? WHERE name=?", (new_phone_number, name))
    conn.commit()
    conn.close()

# Function to delete a specified row
def delete_entry(name):
    conn = sqlite3.connect("phonebook.db")
    cursor = conn.cursor()
    cursor.execute("DELETE FROM Entries WHERE name=?", (name,))
    conn.commit()
    conn.close()

# Main function for user interaction
def main():
    # Create the phonebook database and table
    create_phonebook_db()

    while True:
        print("\nPhonebook Menu:")
        print("1. Add Entry")
        print("2. Look Up Entry")
        print("3. Change Entry")
        print("4. Delete Entry")
        print("5. Exit")

        choice = input("Enter your choice (1-5): ")

        if choice == "1":
            # Add Entry
            name = input("Enter name: ")
            phone_number = input("Enter phone number: ")
            add_entry(name, phone_number)
            print("Entry added successfully.")

        elif choice == "2":
            # Look Up Entry
            name = input("Enter name to look up: ")
            result = lookup_entry(name)
            print(f"Phone number: {result}")

        elif choice == "3":
            # Change Entry
            name = input("Enter name to change phone number: ")
            new_phone_number = input("Enter new phone number: ")
            change_entry(name, new_phone_number)
            print("Phone number changed successfully.")

        elif choice == "4":
            # Delete Entry
            name = input("Enter name to delete: ")
            delete_entry(name)
            print("Entry deleted successfully.")

        elif choice == "5":
            print("Exiting program. Goodbye!")
            break

        else:
            print("Invalid choice. Please enter a number between 1 and 5.")

if __name__ == "__main__":
    main()
```
